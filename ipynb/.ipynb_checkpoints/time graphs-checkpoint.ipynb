```python
%matplotlib inline

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from collections import defaultdict
```

```python
class_name = 'Occupancy'
df = pd.read_csv('../data/training.csv', skipinitialspace=True, na_values='?', 
                 keep_default_na=True)

```

```python
fig, ax = plt.subplots()
ax.boxplot(df['Light'])
ax.set_ylabel('Light')
ax.set_xticks([])
```

```python
df['date'] =  pd.to_datetime(df['date'], format='%Y-%m-%d %H:%M:%S')
df['Time']= df['date'].apply(lambda x:x.time())
df['Date']= df['date'].apply(lambda x:x.date())
df['Time'] =  pd.to_timedelta(str(x) for x in df['Time'])
df['Date'] =  pd.to_datetime(df['Date'], format='%Y-%m-%d')
```

```python
df['Hour']= df['Time'].apply(lambda x:str(x)[7:9])
df['Day']= df['Date'].apply(lambda x:str(x)[8:10])
```

```python
df['Week']=df['Date'].apply(lambda x:x.weekday())
```

```python
df['Hour']=df['Hour'].astype(int)
```
